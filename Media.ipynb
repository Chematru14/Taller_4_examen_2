# Perceptron Multicapa - Funcion Rosenbrock

````python
!pip install neurolab
import numpy as np
import neurolab as nl
import matplotlib.pyplot as plt

lim_inf = -50
lim_sup = 50

# Datos de prueba
ejex = ejey = np.linspace(lim_inf, lim_sup, 50)

def rosenbrock(x, y):
    z = 100 * (x**2 - y)**2 + (x - 1)**2
    return z

#def funcionrandom(x,y):
#    z =
#    return z

x, y = np.meshgrid(ejex, ejey)

z = rosenbrock(x, y)

fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, y, z, cmap='jet')
ax.set_title("Rosenbrock Original")

dim = [lim_inf, lim_sup]


z = z / np.linalg.norm(z)
forma = z.shape


x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
z = z.reshape(-1, 1)

X = np.concatenate((x, y), axis=1)

# Corregir la entrada de la red
net = nl.net.newff([[lim_inf, lim_sup], [lim_inf, lim_sup]], [8, 5, 4, 1])

net.errorf = nl.error.MAE()

error = net.train(X, z, epochs=100, show=10, goal=0.0001)

z_pred = net.sim(X)

plt.figure()
plt.plot(error)
plt.xlabel("Número de épocas")
plt.ylabel("Error")
plt.title("Progreso del error")

# Reformatear los resultados para graficar
z_pred_reshaped = z_pred.reshape(forma)

# Graficar la predicción de la Red Neuronal
ax = fig.add_subplot(122, projection='3d')
x_reshaped = x.reshape(forma)
y_reshaped = y.reshape(forma)
ax.plot_surface(x_reshaped, y_reshaped, z_pred_reshaped, cmap='jet', rstride=1, cstride=1)

# Establecer los límites de los ejes para que coincidan con la función original
ax.set_xlim([lim_inf, lim_sup])
ax.set_ylim([lim_inf, lim_sup])
ax.set_zlim([np.min(z), np.max(z)])

ax.set_title("Predicción de la Red Neuronal")

plt.show()
````

````python
#Librerias
import os
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torchvision import transforms, datasets, models
from torch.utils.data import random_split


#--------PRE-PROCESAMIENTO-----------------------


# Diccionario clasificación
clases = dict({0:"Gato", 1:"Perro",2:"Limon"})
print(clases)

{0: "Gato", 1: "Perro", 2: "Limon"}
dataset_dir = "Gatos"
# Ajustamos las imágenes
transformadas = transforms.Compose([transforms.RandomRotation(45),
transforms.RandomResizedCrop(200),
transforms.RandomHorizontalFlip(),
transforms.ToTensor(),
transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

# Cargamos el dataset de imágenes
dataset = datasets.ImageFolder(dataset_dir, transform=transformadas)
print(dataset)
len_dataset = len(dataset)

# Conjuntos entrenamiento, prueba y validacion
len_entrenamiento = int(len_dataset * 0.7)
len_prueba = int((len_dataset - len_entrenamiento)/2)
len_validacion = len_dataset - len_entrenamiento - len_prueba
entrenamiento, validacion, prueba = random_split(dataset, (len_entrenamiento, len_validacion,len_prueba))

# Dataloader datos de entrenamiento, prueba y validacion
loader_entrenamiento = torch.utils.data.DataLoader(entrenamiento, batch_size=64, shuffle=True)
loader_validacion = torch.utils.data.DataLoader(validacion, batch_size=1, shuffle=True)
loader_prueba = torch.utils.data.DataLoader(prueba, batch_size=1, shuffle=True)

#-------CONSTRUCCION--------------

models.alexnet()

# Modelo personalizado
class CNN(torch.nn.Module):

# Constructor
    def __init__(self):
        super().__init__()
        self.convolutiva = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=(11, 11), stride=(4, 4), padding=(2, 2)),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=0, dilation=1, ceil_mode=False),
            nn.Conv2d(64, 192, kernel_size=(5, 5), stride=(1, 1), padding=(2, 2)),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=0, dilation=1, ceil_mode=False),
            nn.Conv2d(192, 384, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=0, dilation=1, ceil_mode=False),

            nn.AdaptiveAvgPool2d(output_size=(6, 6))
    )
    # Clasificador
        self.clasificador = nn.Sequential(
            nn.Linear(in_features=9216, out_features=4096, bias=True),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5, inplace=False),
            nn.Linear(in_features=4096, out_features=3, bias=True)
    )

    def forward(self, x):
        x = self.convolutiva(x)
        x = torch.flatten(x, start_dim=1, end_dim=-1)
        x = self.clasificador(x)
        return x


# Usamos GPU (si hay)
dispositivo = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instancia  Red Convolucional
cnn = CNN().to(dispositivo)
criterio = nn.CrossEntropyLoss()  # Función de pérdida
optimizador = torch.optim.Adam(cnn.parameters(), lr=0.00001)

def evaluar(modelo, prueba, dispositivo, criterio, mostrar=False):
    # Modo de evaluación
    modelo.eval()
    correctas = 0
    total = 0
    perdida_val = 0.0
    # Datos del data loader
    for imagenes, etiquetas in iter(prueba):
        # Copiamos datos al dispositivo (gpu/cpu)
        imagenes = imagenes.to(dispositivo)
        etiquetas = etiquetas.to(dispositivo)
        salidas = modelo(imagenes)
        # Predicciones
        predichos = torch.max(salidas.data, 1)[1]
        total += len(etiquetas)
        perdida_val += criterio(salidas, etiquetas).data
        # Predicciones correctas
        correctas += (predichos == etiquetas).sum()
    perdida_val = float(perdida_val/float(total))
    perdida_val = round(perdida_val, 4)
    # Exactitud
    exactitud = 100 * correctas / float(total)
    exactitud = round(float(exactitud), 4)
    if mostrar:
        print("Exactitud: {}%".format(exactitud))
    else:
        return exactitud, perdida_val

def entrenar_modelo(modelo, entrenamiento, validacion, criterio, optim, dispositivo, ex_d=100, epocas=100, mostrar=20):
    l_perdida = []
    l_exact = []
    l_perd_val = []
    for epoca in range(epocas + 1):
        perdida_datos = 0.0
        total = 0
        modelo.train()
        # Imágenes y Etiquetas
        for imagenes, etiquetas in iter(entrenamiento):
            # Copiamos datos al dispositivo
            imagenes = imagenes.to(dispositivo)
            etiquetas = etiquetas.to(dispositivo)
            optim.zero_grad()
            # Salida del modelo
            salidas = modelo(imagenes)
            total += len(etiquetas)
            # Pérdida (valor esperado - valor obtenido)
            perdida = criterio(salidas, etiquetas)
            perdida_datos += perdida.data
            # Gradiente (antes 0) usando pérdida estimada
            perdida.backward()
            optim.step()
        # Calculamos la pérdida
        perdida_datos = float(perdida_datos/float(total))
        perdida_datos = round(perdida_datos, 4)
        # Calculamos Exactitud y Pérdida de validación
        exact, perd_val = evaluar(modelo, validacion, dispositivo, criterio)
        # Guardamos (graficos)
        l_perdida.append(perdida_datos)
        l_exact.append(exact)
        l_perd_val.append(perd_val)
        # Si se llega a pérdida aceptada y se pierde exactitud, se detiene entrenamiento
        if ex_d <= exact:
            # Imprimimos
            print("Época: {}\tPérdida entrenamiento: {}\tPerdida validacion: {}\tExactitud validacion: {}".format(epoca, perdida_datos, perd_val, exact))
            break
        if epoca % mostrar == 0:
            print("Época: {}\tPérdida entrenamiento: {}\tPerdida validacion: {}\tExactitud validacion: {}".format(epoca, perdida_datos, perd_val, exact))
    # Convertimos listas a arrays
    l_perdida = np.asarray(l_perdida)
    l_exact = np.asarray(l_exact)
    l_perd_val = np.asarray(l_perd_val)
    l_epoca = np.arange(epoca + 1)

    return l_perdida, l_perd_val, l_exact, l_epoca

'''
#Llamamos la funcion entrenar modelo
perdida, perd_val, exactitud, epoca = entrenar_modelo(cnn, loader_entrenamiento, loader_validacion, criterio, optimizador, dispositivo, ex_d=80, mostrar=1 )

# Escalamos la pérdida de validación
perd_val1 = perd_val*(perdida.mean()/perd_val.mean())
perd_val1

# Función de pérdida
plt.plot(epoca, perdida, color='g')
plt.plot(epoca, perd_val1, color='b')
plt.xlabel("Epoca")
plt.ylabel("Pérdida")
plt.title("Perdida")
plt.legend(["Entrenamiento", "Validacion"])
plt.show()

# Exactitud
plt.plot(epoca, exactitud, color='b')
plt.xlabel("Epoca")
plt.ylabel("Exactitud")
plt.title("Exactitud")
plt.show()
'''
#SOLO LLAMAMOS LA FUNCION ENTRENAR MODELO 1 VEZ

evaluar(cnn, loader_prueba, dispositivo, criterio, mostrar=True)

def mostrar_imagen(loader, k=-1):
    mean = torch.tensor([0.485, 0.456, 0.406])
    std = torch.tensor([0.229, 0.224, 0.225])
    i = 0
    for imagenes, etiquetas in iter(loader):
        for imagen in imagenes:
            i += 1
            # Condición de paro
            if k > 0 and k < i:
                break
            # Revertimos la normalización de la imagen
            img = imagen * std[:,None,None] + mean[:,None,None]
            plt.imshow(img.numpy().transpose(1, 2, 0))
            plt.show()
        # Condición de paro
        if k > 0 and k < i:
            break

mostrar_imagen(loader_prueba, k=3)

def evaluar(modelo, prueba, dispositivo, criterio, mostrar=False, predecir=False, k=-1):
    modelo.eval()
    correctas = 0
    total = 0
    perdida_val = 0.0
    j = 0
    # Datos data loader
    for imagenes, etiquetas in iter(prueba):
        # Copiamos datos al dispositivo (gpu/cpu)
        imagenes = imagenes.to(dispositivo)
        etiquetas = etiquetas.to(dispositivo)
        salidas = modelo(imagenes)
        # Predicciones
        predichos = torch.max(salidas.data, 1)[1]
        total += len(etiquetas)
        # Pérdida
        perdida_val += criterio(salidas, etiquetas).data
        # Predicciones correctas
        correctas += (predichos == etiquetas).sum()
        # Ver imágenes
        if predecir:
            if j < k or k == -1:
                # Imágenes al cpu
                imagenes = imagenes.to("cpu")
                predichos = predichos.to("cpu")
                predichos = predichos.numpy()
                # Índice de la predicción
                i = 0
                mean = torch.tensor([0.485, 0.456, 0.406])
                std = torch.tensor([0.229, 0.224, 0.225])
                for imagen in imagenes:
                    if k > 0:
                        j += 1
                    # Desnormalizamos la imagen
                    img = imagen * std[:,None,None] + mean[:,None,None]
                    plt.imshow(img.numpy().transpose(1, 2, 0))
                    # Predicción imagen
                    prediccion = clases.get(predichos[i])
                    plt.title(prediccion)
                    i += 1
                    plt.show()
    perdida_val = float(perdida_val/float(total))
    perdida_val = round(perdida_val, 4)
    # Exactitud
    exactitud = 100 * correctas / float(total)
    exactitud = round(float(exactitud), 4)
    if mostrar:
        print("Exactitud: {}%".format(exactitud))
    else:
        return exactitud, perdida_val

evaluar(cnn, loader_prueba, dispositivo, criterio, mostrar=True, predecir=True, k=10)

# Guardar modelo (Sólo 1 vez)
#torch.save({"model_state_dict": cnn.state_dict(),"optimizer_state_dict": optimizador.state_dict(),},"cnn_checkpoint.pth")

# Usamos gpu (si hay)
dispositivo = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instanciamos Clase
cnn = CNN().to(dispositivo)
criterio = nn.CrossEntropyLoss()
optimizador = torch.optim.SGD(cnn.parameters(), lr=0.01)
# Cargamos valores guardados
checkpoint = torch.load("cnn_checkpoint.pth")
cnn.load_state_dict(checkpoint["model_state_dict"])
optimizador.load_state_dict(checkpoint["optimizer_state_dict"])
````
